```python
import re

linefix = {}

fname = 'tlg0012.tlg001.newman-eng1.xml'
outfname = 'tlg0012.tlg001.newman-eng1b.xml'
outf = open(outfname,'w')

workbuf = ''

f = open(fname)

curbook = ''
curlnum = 0
for l in f:
    l = re.sub('<fw[^>]*>.+</fw>','',l)
    ## particular to the Newman Iliad
    l = re.sub('<milestone unit="line" ed="grc" n="[0-9]+"/>','',l)
    l = re.sub('\s*[0-9]+(</seg>)','\g<1>',l)
    
    m = re.search('<div type="textpart" subtype="book" n="([^"]+)',l)
    if(m):
        curbook = m[1] + '.'
        print(curbook)
        curlnum = 0
    m = re.search('<div type="textpart" subtype="(argument|text)',l)
    if(m):
        curlnum = 0

    m = re.search('<lb\s+n="([0-9]+)"/>',l)
    if(m):
        curlnum = int(m[1])-1
    #m = re.search('<l n="([^"]+)',l)
    m = re.search('<l n="([0-9]+)"',l)
    if(m):
        curline = '"' + m[1] + '"'
        curcit = curbook + m[1]
        curlnum = curlnum + 1
        realcit = curbook + str(curlnum)
        if(not realcit == curcit):
            print(realcit,curcit)
            linefix[curcit] = realcit
            rightline = '"' + re.sub('^[0-9]+\.','',realcit) + '"'
            l = re.sub(curline,rightline,l)
    workbuf = workbuf + l

for foo in linefix:
    continue
    oldkey = '#n' + foo + '_'
    newkey = '#n' + linefix[foo] + '_'
    workbuf = workbuf.replace(oldkey,newkey)
    

outf.write(workbuf)
outf.close()

        

f.close()
```
