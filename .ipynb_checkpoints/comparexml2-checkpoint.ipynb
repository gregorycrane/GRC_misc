```python
import re

fname = 'suet_holland/suetonius.otho.holland1898-section.xml'
fname2 = re.sub('suet_holland','suet_holland/save',fname)
fname2 = re.sub('1898-','1898a-',fname2)

print(fname,fname2)
```

```python
def dumpfile(fname):
    f = open(fname)
    foutname = re.sub('\.xml','.txt',fname)
    fout = open(foutname,'w')
    wordlist = ''

    for l in f:
        l = re.sub('<[^>]+>',' ',l)
        l = re.sub('\s+','\n',l)
        l = re.sub('\s+$','',l)
        if(l):
            wordlist = wordlist  + l 
        
        

    f.close()
    #print(wordlist)
    fout.write(wordlist)
    fout.close()
    
    
dumpfile(fname)
dumpfile(fname2)
```
