```python
import re

fname = 'suet_holland/suetonius.dom.holland1898-section.xml'
fname2 = re.sub('-section','',fname)

```

```python
def dumpfile(fname):
    f = open(fname)
    foutname = re.sub('\.xml','.txt',fname)
    fout = open(foutname,'w')
    wordlist = ''

    for l in f:
        l = re.sub('<[^>]+>',' ',l)
        l = re.sub('\s+','\n',l)
        l = re.sub('\s+$','',l)
        if(l):
            wordlist = wordlist  + l 
        
        

    f.close()
    #print(wordlist)
    fout.write(wordlist)
    fout.close()
    
    
dumpfile(fname)
dumpfile(fname2)
```
