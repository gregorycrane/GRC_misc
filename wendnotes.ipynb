```python
import re
seenbibs = {}
def addentry(curdict,curent):
    if(curent in curdict):
        curdict[curent] = curdict[curent] + 1
    else:
        curdict[curent] =  1
        
def countwords(fname):
    f = open(fname)
    words = 0
    for l in f:
        l = re.sub(':[0-9][^"]*','',l)
        m = re.search('<bibl n="([^"]+)',l)
        if(m):
            addentry(seenbibs,m[1])
        l = re.sub('<[^>]+>',' ',l)
        words = len(l.split()) + words
    f.close()
    print(fname,words)
    return(words)

reverse_dict = {v: k for k, v in tgnums.items()}

foo = countwords('tlg5012.tlg001.wendel1935-grc2.xml')
print(foo * (330/122))
countwords('tlg5012.tlg001.wendel1935-grc2a.xml')
countwords('apollonius.seaton-oct-grc2.xml')
i = 0
for foo in sorted(seenbibs,key=seenbibs.get,reverse=True):
    i = i + 1
    print(i,foo,seenbibs[foo],reverse_dict[foo])
```
