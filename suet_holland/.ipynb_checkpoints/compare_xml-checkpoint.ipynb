# XML Character-Data Comparison
Compares the text content of two XML files to detect unintended changes introduced by markup additions.

**What it checks:**
- `.text` — character data immediately inside an element (before any child tag)
- `.tail` — character data that follows a closing tag (common in mixed-content documents)

**Report categories:**
| Category | Meaning |
|---|---|
| `CHANGED` | Node exists in both files but text differs (includes similarity %) |
| `ONLY IN A` | Text node exists in the original but is absent in the modified file |
| `ONLY IN B` | Text node was introduced by the markup changes |

## Cell 1 — Imports

```python
import re
from pathlib import Path
from xml.etree import ElementTree as ET
from difflib import SequenceMatcher
from IPython.display import display, HTML
```

## Cell 2 — Configuration
Set your file paths and options here.

```python
# ── Edit these three settings ──────────────────────────────────────────────

FILE_A = "suetonius.aug.holland1898.xml"      # Original XML file (before markup changes)
FILE_B = "suetonius.aug.holland1898-section.xml"      # Modified XML file (after markup changes)

IGNORE_WHITESPACE = False     # Set True to collapse whitespace before comparing

# Optional: write the report to a .txt file as well (set to None to skip)
OUTPUT_FILE = None            # e.g. "report.txt"

# ──────────────────────────────────────────────────────────────────────────
```

## Cell 3 — Core functions

```python
# ── Helpers ────────────────────────────────────────────────────────────────

def normalise_ws(text: str) -> str:
    """Collapse all whitespace runs to a single space and strip ends."""
    return re.sub(r"\s+", " ", text).strip()


def get_element_path(element, root):
    """Build an XPath-like string for an element by walking the tree."""
    def _path(el, parent_map, parts):
        parent = parent_map.get(el)
        if parent is None:
            parts.append(el.tag)
            return
        siblings = [c for c in parent if c.tag == el.tag]
        if len(siblings) > 1:
            idx = siblings.index(el) + 1
            parts.append(f"{el.tag}[{idx}]")
        else:
            parts.append(el.tag)
        _path(parent, parent_map, parts)

    parent_map = {child: parent for parent in root.iter() for child in parent}
    parts = []
    _path(element, parent_map, parts)
    parts.reverse()
    return "/" + "/".join(parts)


def extract_text_nodes(root, ignore_whitespace: bool):
    """
    Walk every element and yield (xpath, text_type, value) tuples where
    text_type is 'text' or 'tail'.
    """
    for el in root.iter():
        path = get_element_path(el, root)
        raw_text = el.text or ""
        text_val = normalise_ws(raw_text) if ignore_whitespace else raw_text
        if text_val:
            yield (path, "text", text_val)
        raw_tail = el.tail or ""
        tail_val = normalise_ws(raw_tail) if ignore_whitespace else raw_tail
        if tail_val:
            yield (path, "tail", tail_val)


def parse_file(filepath: str):
    """Parse an XML file, raising clear errors on failure."""
    try:
        tree = ET.parse(filepath)
        return tree.getroot()
    except ET.ParseError as exc:
        raise ValueError(f"Could not parse '{filepath}': {exc}") from exc
    except FileNotFoundError:
        raise FileNotFoundError(f"File not found: '{filepath}'")


# ── Difference model ───────────────────────────────────────────────────────

class Difference:
    __slots__ = ("xpath", "text_type", "value_a", "value_b", "kind")

    def __init__(self, xpath, text_type, value_a, value_b, kind):
        self.xpath = xpath
        self.text_type = text_type
        self.value_a = value_a
        self.value_b = value_b
        self.kind = kind   # 'changed' | 'only_in_a' | 'only_in_b'

    def similarity(self):
        if self.kind != "changed":
            return None
        return SequenceMatcher(None, self.value_a, self.value_b).ratio()


# ── Comparison ─────────────────────────────────────────────────────────────

def compare(nodes_a, nodes_b):
    """Compare two ordered lists of (xpath, text_type, value) tuples."""
    map_a, map_b = {}, {}
    for xpath, ttype, val in nodes_a:
        map_a.setdefault((xpath, ttype), []).append(val)
    for xpath, ttype, val in nodes_b:
        map_b.setdefault((xpath, ttype), []).append(val)

    diffs = []
    for key in sorted(set(map_a) | set(map_b)):
        xpath, ttype = key
        vals_a = map_a.get(key, [])
        vals_b = map_b.get(key, [])
        for i, (va, vb) in enumerate(zip(vals_a, vals_b)):
            if va != vb:
                indexed = f"{xpath}[occurrence {i+1}]" if len(vals_a) > 1 else xpath
                diffs.append(Difference(indexed, ttype, va, vb, "changed"))
        for va in vals_a[len(vals_b):]:
            diffs.append(Difference(xpath, ttype, va, "", "only_in_a"))
        for vb in vals_b[len(vals_a):]:
            diffs.append(Difference(xpath, ttype, "", vb, "only_in_b"))
    return diffs


print("✓ Functions defined.")
```

## Cell 4 — Run the comparison

```python
root_a = parse_file(FILE_A)
root_b = parse_file(FILE_B)

nodes_a = list(extract_text_nodes(root_a, IGNORE_WHITESPACE))
nodes_b = list(extract_text_nodes(root_b, IGNORE_WHITESPACE))

diffs = compare(nodes_a, nodes_b)

changed = [d for d in diffs if d.kind == "changed"]
only_a  = [d for d in diffs if d.kind == "only_in_a"]
only_b  = [d for d in diffs if d.kind == "only_in_b"]

print(f"Files parsed successfully.")
print(f"  Text nodes in A : {len(nodes_a)}")
print(f"  Text nodes in B : {len(nodes_b)}")
print(f"  Differences     : {len(diffs)} ({len(changed)} changed, {len(only_a)} only-in-A, {len(only_b)} only-in-B)")
```

## Cell 5 — Display HTML report

```python
def _esc(s):
    return s.replace("&", "&amp;").replace("<", "&lt;").replace(">", "&gt;")

def truncate(s, max_len=1200):
    return s if len(s) <= max_len else s[:max_len] + "…"

def render_html_report(diffs, file_a, file_b, nodes_a, nodes_b):
    changed = [d for d in diffs if d.kind == "changed"]
    only_a  = [d for d in diffs if d.kind == "only_in_a"]
    only_b  = [d for d in diffs if d.kind == "only_in_b"]
    total   = max(len(nodes_a), len(nodes_b))

    # ── Badge colour ──
    badge_colour = "#2e7d32" if not diffs else "#c62828"
    badge_text   = "✓ No differences" if not diffs else f"✗ {len(diffs)} difference(s) found"

    html = f"""
    <style>
      .xml-report {{ font-family: monospace; font-size: 13px; }}
      .xml-report h2 {{ font-family: sans-serif; border-bottom: 2px solid #555; padding-bottom: 4px; }}
      .xml-report h3 {{ font-family: sans-serif; margin-top: 1.5em; }}
      .badge {{ display:inline-block; padding:4px 12px; border-radius:4px;
                color:#fff; font-weight:bold; font-family:sans-serif;
                background:{badge_colour}; margin-bottom:12px; }}
      .summary-table {{ border-collapse:collapse; margin-bottom:1em; }}
      .summary-table td {{ padding:3px 16px 3px 0; }}
      .diff-card {{ border:1px solid #ddd; border-radius:4px; padding:10px 14px;
                    margin-bottom:10px; background:#fafafa; }}
      .diff-card .xpath {{ color:#1565c0; font-weight:bold; }}
      .diff-card .label {{ color:#555; min-width:80px; display:inline-block; }}
      .val-a {{ color:#c62828; }}
      .val-b {{ color:#2e7d32; }}
      .val-single {{ color:#6a1b9a; }}
      .sim {{ color:#888; font-size:11px; }}
      .section-changed h3 {{ color:#c62828; }}
      .section-only-a  h3 {{ color:#e65100; }}
      .section-only-b  h3 {{ color:#1565c0; }}
    </style>
    <div class="xml-report">
      <h2>XML Character-Data Comparison Report</h2>
      <div class="badge">{badge_text}</div>
      <table class="summary-table">
        <tr><td><b>File A</b></td><td>{_esc(file_a)}</td></tr>
        <tr><td><b>File B</b></td><td>{_esc(file_b)}</td></tr>
        <tr><td><b>Text nodes checked</b></td><td>{total}</td></tr>
        <tr><td><b>Changed</b></td><td>{len(changed)}</td></tr>
        <tr><td><b>Only in A</b></td><td>{len(only_a)}</td></tr>
        <tr><td><b>Only in B</b></td><td>{len(only_b)}</td></tr>
      </table>
    """

    if not diffs:
        html += "<p>No character-data differences detected.</p></div>"
        return html

    # ── Changed ──
    if changed:
        html += '<div class="section-changed"><h3>🔄 Changed — content differs between A and B</h3>'
        for i, d in enumerate(changed, 1):
            sim = d.similarity()
            sim_str = f'<span class="sim">  similarity: {sim:.0%}</span>' if sim is not None else ""
            html += f"""
            <div class="diff-card">
              <div><span class="label">#{i} XPath</span>
                   <span class="xpath">{_esc(d.xpath)}</span>
                   [{d.text_type}]{sim_str}</div>
              <div><span class="label">File A</span>
                   <span class="val-a">{_esc(truncate(repr(d.value_a)))}</span></div>
              <div><span class="label">File B</span>
                   <span class="val-b">{_esc(truncate(repr(d.value_b)))}</span></div>
            </div>"""
        html += "</div>"

    # ── Only in A ──
    if only_a:
        html += '<div class="section-only-a"><h3>⬅ Only in A — missing from B</h3>'
        for i, d in enumerate(only_a, 1):
            html += f"""
            <div class="diff-card">
              <div><span class="label">#{i} XPath</span>
                   <span class="xpath">{_esc(d.xpath)}</span> [{d.text_type}]</div>
              <div><span class="label">Value</span>
                   <span class="val-single">{_esc(truncate(repr(d.value_a)))}</span></div>
            </div>"""
        html += "</div>"

    # ── Only in B ──
    if only_b:
        html += '<div class="section-only-b"><h3>➡ Only in B — not present in A</h3>'
        for i, d in enumerate(only_b, 1):
            html += f"""
            <div class="diff-card">
              <div><span class="label">#{i} XPath</span>
                   <span class="xpath">{_esc(d.xpath)}</span> [{d.text_type}]</div>
              <div><span class="label">Value</span>
                   <span class="val-single">{_esc(truncate(repr(d.value_b)))}</span></div>
            </div>"""
        html += "</div>"

    html += "</div>"  # .xml-report
    return html


display(HTML(render_html_report(diffs, FILE_A, FILE_B, nodes_a, nodes_b)))
```

## Cell 6 — (Optional) Save plain-text report

```python
def build_text_report(diffs, file_a, file_b, nodes_a, nodes_b):
    sep = "─" * 72
    changed = [d for d in diffs if d.kind == "changed"]
    only_a  = [d for d in diffs if d.kind == "only_in_a"]
    only_b  = [d for d in diffs if d.kind == "only_in_b"]
    total   = max(len(nodes_a), len(nodes_b))

    lines = [
        "XML CHARACTER-DATA COMPARISON REPORT", sep,
        f"  File A             : {file_a}",
        f"  File B             : {file_b}", sep,
        f"  Text nodes checked : {total}",
        f"  Differences found  : {len(diffs)}",
        f"    • Changed        : {len(changed)}",
        f"    • Only in A      : {len(only_a)}",
        f"    • Only in B      : {len(only_b)}", sep,
    ]

    if not diffs:
        lines.append("\n✓ No character-data differences detected.\n")
        return "\n".join(lines)

    if changed:
        lines.append("\n[CHANGED]\n")
        for i, d in enumerate(changed, 1):
            sim = d.similarity()
            sim_str = f"  (similarity: {sim:.0%})" if sim is not None else ""
            lines += [
                f"  {i}. XPath  : {d.xpath}  [{d.text_type}]{sim_str}",
                f"     File A  : {truncate(repr(d.value_a))}",
                f"     File B  : {truncate(repr(d.value_b))}", "",
            ]

    if only_a:
        lines.append("[ONLY IN A]\n")
        for i, d in enumerate(only_a, 1):
            lines += [f"  {i}. XPath  : {d.xpath}  [{d.text_type}]",
                      f"     Value   : {truncate(repr(d.value_a))}", ""]

    if only_b:
        lines.append("[ONLY IN B]\n")
        for i, d in enumerate(only_b, 1):
            lines += [f"  {i}. XPath  : {d.xpath}  [{d.text_type}]",
                      f"     Value   : {truncate(repr(d.value_b))}", ""]

    lines += [sep, f"Total differences: {len(diffs)}", ""]
    return "\n".join(lines)


if OUTPUT_FILE:
    report_text = build_text_report(diffs, FILE_A, FILE_B, nodes_a, nodes_b)
    Path(OUTPUT_FILE).write_text(report_text, encoding="utf-8")
    print(f"Plain-text report written to: {OUTPUT_FILE}")
else:
    print("OUTPUT_FILE is not set — skipping file save. Set it in Cell 2 to save a report.")
```
